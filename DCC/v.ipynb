### The big issue now is to write a function that is going to maintain the geojson format and simply find and add all the dataframe data back into the original geojson file correctly. This requires research in the udnerstanding of what exactly a geojson file is and why it needs to be created a certain way.

```python
country_corrections_csv_gj
```

```python
#This functio take a dataframe and parses the desired geojson features of interest back into our downloaded geojson file and saves it
def update_df(key,value):
    df=value
    #pdb.set_trace()
    csv_country_List = df['Country'].values
    
    #First do a loop to check for countries with naming conflicts
    for country in csv_country_List:
        if country in country_corrections_csv_gj.keys():#Check to see if the country is in the list of countries with naming conflicts so it can be renamed
            conditional_index = df.loc[ df['Country'] == country].index #Get the index number of the country
            df.loc[conditional_index , ['Country']] = country_corrections_csv_gj[country]#Rename the country to be the same as that would be found in the geojson file which contains all known countries 

    #Start looping through csv/DataFrame countries to insert the Kaggle data into the already existing geojson file in the correct format so it can later be used with Leaflet
    for country in csv_country_List:
        if country in GJ_country_dict.keys():
            #Strip all Happiness_Rank data as well as all other features
            Happiness_RankAdd = {}
            Happiness_Rank = df[df['Country'] == country]["Happiness_Rank"].values[0]
            Happiness_RankAdd =  {'Happiness_Rank': int(Happiness_Rank)}

            Happiness_ScoreAdd = {}
            Happiness_Score = df[df['Country'] == country]["Happiness_Score"].values[0]
            Happiness_ScoreAdd =  {'Happiness_Score': float(Happiness_Score)}
            
            HealthAdd = {}
            Health = df[df['Country'] == country]["Health"].values[0]
            HealthAdd =  {'Health': Health}
            
            FamilyAdd = {}
            Family = df[df['Country'] == country]["Family"].values[0]
            FamilyAdd =  {'Family': Family}
            
            TrustAdd = {}
            Trust = df[df['Country'] == country]["Trust"].values[0]
            TrustAdd =  {'Trust': Trust}
            
            FreedomAdd = {}
            Freedom = df[df['Country'] == country]["Freedom"].values[0]
            FreedomAdd =  {'Freedom': Freedom}
            
            #Check again if the country is indeed valid then update the new properties
            #Use the Pandas "update" function to add on a new feature/dictionary entry to the geojson 
            if  gj['features'][GJ_country_dict[country]]:#Just another check to make sure the country exists
                gj['features'][GJ_country_dict[country]]["properties"].update(Happiness_RankAdd)
                gj['features'][GJ_country_dict[country]]["properties"].update(Happiness_ScoreAdd)
                gj['features'][GJ_country_dict[country]]["properties"].update(HealthAdd)
                gj['features'][GJ_country_dict[country]]["properties"].update(FamilyAdd)
                gj['features'][GJ_country_dict[country]]["properties"].update(TrustAdd)
                gj['features'][GJ_country_dict[country]]["properties"].update(FreedomAdd)

        else:
            print(f'{country} Does not exist and is ignored' ) # These countries do not exist
    
    with open('./idklol/data/world_happiness_'+key+'.geojson', 'w') as f:
        dump(gj, f)
        f.close()
```
